```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.datasets import load_wine
```

```python
wine=load_wine()
```

```python
data=wine.data
target=wine.target
target_name=wine.target_names

df=pd.DataFrame(data,columns=wine.feature_names)
df['target']=target
```

```python
#PCA Technique
pca=PCA(n_components=2)
pca_result=pca.fit_transform(data)

df['PCA1']=pca_result[:,0]
df['PCA2']=pca_result[:,1]

plt.figure(figsize=(10,6))
sns.scatterplot(x='PCA1',y='PCA2',hue='target',palette='viridis',data=df)
plt.title('PCA of wine Dataset')
plt.show()
```

```python
#t-SNE technique

tsne=TSNE(n_components=2,random_state=42)
tsne_result=tsne.fit_transform(data)

df['TSNE1']=tsne_result[:,0]
df['TSNE2']=tsne_result[:,1]

plt.figure(figsize=(10,6))
sns.scatterplot(x='TSNE1',y='TSNE2',hue='target',palette='viridis',data=df)
plt.title('t-SNE of wine Dataset')
plt.show()
```
